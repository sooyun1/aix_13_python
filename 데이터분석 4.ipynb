```python
# parse_dates는 date의 object를 datetime로 바꿔줌
ebola = pd.read_csv(
    './data/country_timeseries.csv',
    parse_dates=['Date']
)
ebola.info()
```

```python
# 바이러스 발병난지 얼마나 지났는지
ebola['outbreak_d'] = ebola['Date'] - ebola['Date'].min() #최소값
print(ebola[['Date','Day','outbreak_d']])
```

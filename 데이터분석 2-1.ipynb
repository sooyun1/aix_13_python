```python
#전채 탑승객 중 최고 나이, 최소 나이
print(df['Age'].max(), df['Age'].min())

#성별 최고 나이, 최소 나이
df.groupby('Sex')['Age'].max(), df.groupby('Sex')['Age'].min()
```

```python
#선실 등급별 최고, 최소, 평균 나이

df.groupby('Pclass')['Age'].max()
df.groupby('Pclass')['Age'].min()
df.groupby('Pclass')['Age'].mean()
```

```python
#성별, 선실 등급별 생존율
df.groupby(['Sex','Pclass'])['Survived'].mean()
```

```python
#Age의 결측치를 채우기(평균값으로)
b1 = df['Age'].isna()
df[b1]
```

```python
age_mean = df['Age'].mean()
df['Age'].fillna(age_mean) #모든 나이가 채워진 복사본이 결과값으로 나옴. 원본 수정 안 됨.
# #원본 수정법
# 1. inplace=True 부여
# 2. 원본칼럼을 덮어쓴다.
df['Age'] = df['Age'].fillna(age_mean)
```

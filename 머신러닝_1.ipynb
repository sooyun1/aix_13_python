```python
bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0, 31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0, 35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0, 500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0, 700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]
```

```python
import matplotlib.pyplot as plt

plt.scatter(bream_length, bream_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
```

```python
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
length = bream_length+smelt_length
weight = bream_weight+smelt_weight
```

```python
len(length), len(weight)
```

```python
fish_data = [[l, j] for l, j in zip(length, weight)]
fish_data
```

```python
fish_target = [1]*35 + [0]*14
print(fish_target)
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
kn.fit(fish_data, fish_target) # 학습 (입력, 출력을 학습)
```

```python
kn.predict([[29.0, 363.0]]) #2차원 배열로 전달
```

```python
kn.predict([[10.5, 7.5]])
```

```python
kn.predict([[30,600]]) # 없는 데이터(이웃 하는 값의 갯수를 보고 비중이 많은 값으로 출력)
```

```python
import pandas as pd
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/refs/heads/master/titanic.csv'
titanic = pd.read_csv(url)
```

```python
titanic.info()
```

```python
# 머신러닝은 입력값에 결측치가 있으면 오류 발생
input = titanic[['Pclass', 'Age']]
input.isna().sum()
```

```python
input
```

```python
input.plot.scatter(x='Age', y='Pclass')
```

```python
# Survived 종류에 따라 찍어보기
titanic.query('Survived == 0').plot.scatter(x='Age', y='Pclass')
titanic.query('Survived == 1').plot.scatter(x='Age', y='Pclass')
```

```python
# seaborn은 같은 조건에 여러개 결과를 같이 출력할 수 있음
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(data=titanic, x='Age', y='Pclass', hue='Survived', palette={0:'red', 1:'blue'}) # hue:그룹나누기

plt.show()
```

```python
# Age를 평균값으로 채움
input['Age'] = input['Age'].fillna(input['Age'].mean())
input.isna().sum()
```

```python
survived = titanic['Survived'] # 예측하고자 하는 것
```

```python
# KNeighborsClassifier 최근접 이웃 알고리즘 로딩
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
kn.fit(input, survived)
```

```python
kn.predict([[1, 27]])
```

```python
# 새로운 입력 특성으로 Pclass, Age, Fare 선택
input_new = titanic[['Pclass', 'Age', 'Fare']]

# Age의 결측치를 평균으로 채움 (Fare는 결측치가 없는 것을 확인)
input_new['Age'] = input_new['Age'].fillna(input_new['Age'].mean())

# 결측치 확인
input_new.isna().sum()
```

```python
# 근접이웃 모델 재학습
from sklearn.neighbors import KNeighborsClassifier
kn_new = KNeighborsClassifier()
kn_new.fit(input_new, survived)
```

```python
# 새로운 데이터로 예측
prediction_example = kn_new.predict([[2,27,15]])
print(f"[Pclass=2, Age=27, Fare=15] 데이터에 대한 예측 결과: {prediction_example}")
```

```python
input
```

```python
# kn.score( 훈련데이터, 답데이터 )
# 훈련데이터가 예측한 값과 답 데이터를 비교, 점수를 냄
test_data = [[3,22],[1,38],[3,26],[1,35]]
pred = kn.predict(test_data)
result = (pred == survived[:4])
print(result)
print(result.mean())

# 형태 : kn.score( 훈련데이터, 답데이터 )
print(kn.score(test_data, survived[:4])) # 한방에 처리 (사실 print도 안 써도 됨)

# kn.score를 쓰면 좋은 점 : 코드 길이가 짧아(간편)짐. 속도는 동일함.
```

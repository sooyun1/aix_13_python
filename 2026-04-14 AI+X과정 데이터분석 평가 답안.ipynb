##  [문제 1] BMI 계산 및 분포 분석

* 신장과 체중을 이용하여 BMI 컬럼을 생성하시오.
* BMI 분포를 히스토그램으로 시각화하시오.
* BMI 분포의 특징을 간단히 서술하시오.

```python
df['신장_m'] = df['신장(5cm단위)'] / 100
df['BMI'] = df['체중(5kg단위)'] / (df['신장_m'] ** 2)

# 그래프
plt.hist(df['BMI'].dropna(), bins=30)
plt.title("BMI 분포")
plt.xlabel("BMI")
plt.ylabel("Count")
plt.show()
```

##  [문제 2] 비만 분류 및 비율 분석

* BMI 기준으로 다음과 같이 분류하시오.

  * 저체중 / 정상 / 과체중 / 비만
* 각 그룹의 인원수를 구하시오.
* 막대그래프로 시각화하시오.
* 가장 높은 비율의 그룹을 분석하시오.

```python
def bmi_category(bmi):
    if bmi < 18.5:
        return '저체중'
    elif bmi < 23:
        return '정상'
    elif bmi < 25:
        return '과체중'
    else:
        return '비만'

df['비만여부'] = df['BMI'].apply(bmi_category)

# 그래프
df['비만여부'].value_counts().plot(kind='bar')
plt.title("비만 분류")
plt.show()
```

##  [문제 3] 복부비만 분석

* 성별 기준 허리둘레로 복부비만 여부를 판단하시오.
* 성별에 따른 복부비만 비율을 계산하시오.
* 막대그래프로 시각화하시오.
* 남녀 차이를 해석하시오.

```python

def abdominal_obesity(row):
    if row['성별코드'] == 1:
        return row['허리둘레'] >= 90
    else:
        return row['허리둘레'] >= 85

df['복부비만'] = df.apply(abdominal_obesity, axis=1)

# 그래프
df.groupby('성별코드명')['복부비만'].mean().plot(kind='bar')
plt.title("성별에 따른 복부비만")
plt.show()

#남자가 복부비만이 많다
```

##  [문제 4] 고혈압 위험군 분석

* 고혈압 기준을 적용하여 위험군을 분류하시오.
* 전체 대비 고혈압 위험군 비율을 구하시오.
* 그래프로 시각화하시오.
* 결과를 해석하시오.

```python

df['고혈압위험군'] = (
    (df['수축기혈압'] >= 140) | (df['이완기혈압'] >= 90)
)

# 그래프
df['고혈압위험군'].value_counts().plot(kind='bar')
plt.title("고혈압 위험군 비율")
plt.show()


# 고혈압 위험군 아닌사람이 많다.
```

##  [문제 5] 혈당 상태 분석

* 공복혈당 기준으로 3단계로 분류하시오.
* 각 그룹의 인원수를 구하시오.
* 막대그래프로 시각화하시오.
* 당뇨 위험군 비율을 해석하시오.

```python

def sugar(x):
    if x < 100:
        return '정상'
    elif x < 126:
        return '위험'
    else:
        return '당뇨'

df['혈당상태'] = df['식전혈당(공복혈당)'].apply(sugar)

df['혈당상태'].value_counts().plot(kind='bar')
plt.title("Glucose")
plt.show()

# 당뇨전단계 사람들이 제법있다.
```

##  [문제 8] 결측치 처리 비교 분석

* 콜레스테롤 관련 컬럼의 결측치를

  1. 평균값으로 대체
  2. 결측치 제거
     두 가지 방법으로 처리하시오.
* 각 방법에 따른 평균값을 비교하는 그래프를 그리시오.
* 어떤 방법이 더 적절한지 설명하시오.

```python

cols = ['총콜레스테롤','HDL콜레스테롤','LDL콜레스테롤']

df_fill = df.copy()
for c in cols:
    df_fill[c].fillna(df_fill[c].mean(), inplace=True)

df_drop = df.dropna(subset=cols)

df_fill[cols].mean().plot(kind='bar')
plt.title("Fillna")
plt.show()

df_drop[cols].mean().plot(kind='bar')
plt.title("Dropna")
plt.show()
```

##  [문제 10] 건강 위험 점수 모델링

* 다음 기준으로 건강 위험 점수를 계산하시오.

  * 비만 / 고혈압 / 당뇨 / 흡연 / 음주
* 점수에 따라 그룹을 분류하시오.

  * 건강 / 주의 / 고위험
* 그룹별 인원 비율을 구하고 그래프로 시각화하시오.
* 고위험군 특징을 분석하시오.

```python

df['비만'] = df['BMI'] >= 25
df['고혈압'] = (df['수축기혈압'] >= 140) | (df['이완기혈압'] >= 90)
df['당뇨'] = df['식전혈당(공복혈당)'] >= 126
df['흡연'] = df['흡연상태'] == 1
df['음주'] = df['음주여부'] == 1

df['점수'] = df[['비만','고혈압','당뇨','흡연','음주']].sum(axis=1)

df['점수'].value_counts().plot(kind='bar')
plt.title("건강 위험 점수")
plt.show()
```

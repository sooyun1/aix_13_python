```python
import folium
# folium.Map(
#     location=[위도, 경도], 
#     zoom_start=배율, 
#     width='가로크기', 
#     height='세로크기')
# 위도:37.50006, 경도:127.0355806, 배율:17 설정한 지도
map_folium = folium.Map(
    location=[37.570529, 126.985304], 
    zoom_start=17)
map_folium
```

```python
# Marker()

# location 속성으로 지정한 위치에 마커를 생성
# folium.Marker(location=[위도, 경도])

# 위도: 37.50006, 경도: 127.0355806, 배율:17 지도 객체 생성
map_folium = folium.Map(location=[37.570529, 126.985304], zoom_start=17)

# 표시할 문자열:'MBC컴퓨터학원', 최대 너비: 200 팝업 객체 생성
popup = folium.Popup("MBC컴퓨터학원", max_width=200)

# 위도: 37.50006, 경도: 127.0355806, popup:위에 생성한 popup객체를 map_folium 지도에 추가
folium.Marker(location=[37.570529, 126.985304], popup=popup).add_to(map_folium)
map_folium
```

```python
# CustomIcon import
from folium.features import CustomIcon
# 이미지 경로 등록
icon_img = "./data/batman.png"
icon = CustomIcon(icon_img, icon_size=(45,45))

seoul_folium = folium.Map(
    location=[37.56652479270633, 126.9779129116262], 
    zoom_start=14)

popup1 = folium.Popup("서울특별시청", max_width=200)
popup2 = folium.Popup("덕수궁", max_width=200)
popup3 = folium.Popup("서울지방경찰청", max_width=200)
popup4 = folium.Popup("광화문", max_width=200)
popup5 = folium.Popup("영풍문고-종로본점", max_width=200)

# 별 모양 아이콘
folium.Marker(location=[37.56652479270633, 126.9779129116262], popup=popup1, icon=folium.Icon(color="red", icon="star")).add_to(seoul_folium)
# 체크 모양 아이콘
folium.Marker(location=[37.56585320879458, 126.97514572197248], popup=popup2, icon=folium.Icon(color="pink", icon="ok")).add_to(seoul_folium)
# 사람 모양 아이콘
folium.Marker(location=[37.57504045678939, 126.97195576938117], popup=popup3, icon=folium.Icon(color="orange", icon="user")).add_to(seoul_folium)
# 위에서 만든 배트맨 아이콘
folium.Marker(location=[37.57617987638928, 126.97694467807803], popup=popup4, icon=icon).add_to(seoul_folium)
# 기본 아이콘
folium.Marker(location=[37.56999947440024, 126.98228391170936], popup=popup5).add_to(seoul_folium)

seoul_folium
```

```python
# 서울시 종로구 카페의 위치를 지도에 표시하기
df.query('상권업종소분류명 == "카페" and 시군구명=="종로구" ')
```

```python
# 서울시 종로구 카페의 위치를 지도에 표시하기
jr_cafe = df.query('상권업종소분류명 == "카페" and 시군구명=="종로구" ')

map_folium = folium.Map(
    location=[37.570529, 126.985304], 
    zoom_start=17)

# 행단위 추출
for _ ,row in jr_cafe.iterrows(): 
    # 해당 행의 시설명을 문자열로 등록
    popup = folium.Popup(
        row["상호명"], 
        max_width=200,
        # show=True  #계속보임
    )
    # 해당 행의 위도와 경도에 마커 등록
    folium.Marker(
        location=[row["위도"], row["경도"]], 
        popup=popup).add_to(map_folium)
map_folium
```

```python
# 서울시 종로구 카페의 위치를 지도에 표시하기
jr_cafe = df.query('상권업종소분류명 == "카페" and 시군구명=="종로구" ')

map_folium = folium.Map(
    location=[37.570529, 126.985304], 
    zoom_start=17)

from folium.plugins import MarkerCluster
# MarkerCluster 객체 생성
mc = MarkerCluster().add_to(map_folium)

# 행단위 추출
for _ ,row in jr_cafe.iterrows(): 
    # 해당 행의 시설명을 문자열로 등록
    popup = folium.Popup(
        row["상호명"], 
        max_width=200,
        # show=True  #계속보임
    )
    # 해당 행의 위도와 경도에 마커 등록
    mc.add_child(
    folium.Marker(
        location=[row["위도"], row["경도"]], 
        popup=popup).add_to(map_folium)
    )
map_folium
```

```python
# 데이터 추출 (행, 열)
# df['칼럼명'] 칼럼 추출
titanic['Survived']

# df[['칼럼명','칼럼명]]
titanic[['Survived','Age']]

# df[ bool(조건식) ] 조건에 맞는 행 추출
b1 = titanic['Age'] < 20 #이걸 많이 씀
b1 = titanic.Age < 20
titanic[ b1 ] # 20세이하 행 추출

# df[ 시작행인덱스:끝행인덱스:step ] 행 추출
titanic[ 3:6 ] # 3~6미만 행
```

```python
# df.loc[행인덱스, 열인덱스]
titanic.loc[1, 'Age']
titanic.loc[1] #행전체

# df.loc[행인덱스, :] 행전체
titanic.loc[1, :]

# df.loc[:, 열인덱스] 열전체
# df.loc[행인덱스,행인덱스] 열인덱스]
# df.loc[행인덱스,[행인덱스, 열인덱스]]
```

```python
# df.query( '칼럼명 비교연산자 값' ) 행 추출 (훨씬 많이 쓰임)
titanic.query('Age < 20')
titanic.query('Pclass in [1,2]')
titanic.query('Pclass == 1 and Sex == "male"')
# titanic.query('Name.str.contains("Mr")')
```

```python
# 칼럼 추가
def to_age_type(x):
    if x < 13: return "child"
    if x < 30: return "youth"
    if x < 50: return "middle-age"
    return "senior"
titanic['age_type'] = titanic['Age'].apply(to_age_type) #나이대
titanic.head()
```

```python
# 그룹별 통계
# df.groupby()[칼럼명].그룹함수
# df.groupby()[칼럼명].agg(['그룹함수'])

# 성별 탑승객수
titanic.groupby('Sex')['Sex'].count()

# 성별 생존율
titanic.groupby('Sex')['Survived'].mean()

# 성별 객실등급별 생존율
titanic.groupby(['Sex','Pclass'])['Survived'].mean()

 #함수 여러개 동시에 실행 가능
titanic.groupby('Sex')['Survived'].agg(['mean','count'])

#칼럼별 다른 통계적용 (딕셔너리 형태로 칼럼의 칼럼을 지정 가능)
titanic.groupby('Sex').agg(
    {'Survived':['mean','count'],
     'Age':'mean'}
)
```

```python
import pandas as pd

scientists = pd.DataFrame(
    data={
        "Occupation": ["Chemist", "Statistician"],
        "Born": ["1920-07-25", "1876-06-13"],
        "Died": ["1958-04-16", "1937-10-16"],
        "Age": [37, 61],
    },
    index=["Rosaline Franklin", "William Gosset"],
    columns=["Occupation", "Born", "Died", "Age"],
)
print(scientists)
```

```python
scientists.info()
```

```python
scientists['born_dt'] = pd.to_datetime( scientists['Born'],format='%Y-%m-%d' )
```

```python
scientists['died_dt'] = pd.to_datetime( scientists['Died'],format='%Y-%m-%d' )
```

```python
scientists.info()
```

```python
#생존날짜 계산
scientists['age_days'] = scientists['died_dt']-scientists['born_dt']
```

```python
scientists
```

```python
scientists.info()       #datetime칼럼만 갖고 있는 속성 .dt(시리즈 내에 있는 값에 반복적으로 계산해주는 변수)
```

```python
import numpy as np

scientists['age_years'] = (scientists['age_days'].dt.days/365).apply(np.floor) #버림(소수점 이하) 반복적으로 처리:apply
```

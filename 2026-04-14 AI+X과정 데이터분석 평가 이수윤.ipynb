##  [문제 1] BMI 계산 및 분포 분석

* 신장과 체중을 이용하여 BMI 컬럼을 생성하시오.
* BMI 분포를 히스토그램으로 시각화하시오.
* BMI 분포의 특징을 간단히 서술하시오.

```python
import pandas as pd
file = './data/국민건강보험공단_건강검진정보_2023.CSV'
df = pd.read_csv(file, encoding='cp949')
```

#### 신장과 체중을 이용하여 BMI 컬럼을 생성

```python
import numpy as np

df['BMI'] = df['체중(5kg단위)'] / ((df['신장(5cm단위)'] / 100) ** 2)
df[['신장(5cm단위)', '체중(5kg단위)', 'BMI']].head()
df
```

#### BMI 분포를 히스토그램으로 시각화

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.hist(df['BMI'], bins=50, edgecolor='black')

plt.title('BMI Distribution')
plt.xlabel('BMI')
plt.ylabel('Frequency')

plt.show()
```

```python
# 씨본

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.histplot(df['BMI'], bins=50, kde=True)

plt.title('BMI Distribution (with KDE)')
plt.xlabel('BMI')
plt.ylabel('Count')

plt.show()
```

#### - BMI 분포의 특징을 간단히 서술
  - 대부분은 정상~과체중 초반에 몰려 있고, 일부 고BMI가 존재
  - 최빈 구간 : 22~27
  - 비만 인구 비중 존재 (25 이상), 저체중 (BMI < 18.5)는 매우 적음
  - 이상치 존재 가능성 : BMI 40~60 구간에도 일부 값 존재 -> 극단적 비만

#### BMI 기준으로 다음과 같이 분류 (저체중 / 정상 / 과체중 / 비만)

* 막대그래프로 시각화하시오.

#### 성별 기준 허리둘레로 복부비만 여부를 판단

```python
import numpy as np

df['복부비만'] = np.where(
    ((df['성별코드'] == 1) & (df['허리둘레'] >= 90)) |
    ((df['성별코드'] == 2) & (df['허리둘레'] >= 85)),
    '비만',
    '정상'
)
df['복부비만'].value_counts()
```

#### 성별에 따른 복부비만 비율을 계산

```python
df.groupby('성별코드')['복부비만'].value_counts(normalize=True) * 100
```

#### 막대그래프로 시각화

```python
import pandas as pd

ratio = pd.crosstab(df['성별코드'], df['복부비만'], normalize='index')

ratio.plot(kind='bar', stacked=True, figsize=(8,5), colormap='Set2')

plt.title('성별 복부비만 비율')
plt.xlabel('성별코드 (1=남성, 2=여성)')
plt.ylabel('비율')

plt.legend(title='복부비만 여부')
plt.show()
```

##  [문제 4] 고혈압 위험군 분석

* 고혈압 기준을 적용하여 위험군을 분류하시오.
* 전체 대비 고혈압 위험군 비율을 구하시오.
* 그래프로 시각화하시오.
* 결과를 해석하시오.

#### 고혈압 기준을 적용하여 위험군을 분류

```python
import numpy as np

df['혈압구간'] = np.select(
    [
        (df['수축기혈압'] >= 140) | (df['이완기혈압'] >= 90),
        (df['수축기혈압'] >= 120) | (df['이완기혈압'] >= 80),
        (df['수축기혈압'] < 120) & (df['이완기혈압'] < 80)
    ],
    [
        '고혈압',
        '전단계',
        '정상'
    ],
    default='기타'
)

df['혈압구간'].value_counts()
```

#### 전체 대비 고혈압 위험군 비율

```python
df['혈압구간'].value_counts(normalize=True) * 100
```

```python
risk_ratio = df['혈압구간'].isin(['전단계', '고혈압']).mean() * 100
print(risk_ratio)
```

```python
total = len(df)

normal = (df['혈압구간'] == '정상').sum() / total * 100
pre = (df['혈압구간'] == '전단계').sum() / total * 100
hyper = (df['혈압구간'] == '고혈압').sum() / total * 100

print(f'정상: {normal:.1f}%')
print(f'전단계: {pre:.1f}%')
print(f'고혈압: {hyper:.1f}%')
```

#### 그래프로 시각화

```python
import matplotlib.pyplot as plt

risk = df['혈압구간'].isin(['전단계', '고혈압']).mean() * 100
normal = 100 - risk

labels = ['정상', '위험군']
values = [normal, risk]

plt.figure(figsize=(6,6))
plt.bar(labels, values, color=['green','red'], edgecolor='black')

plt.title('고혈압 위험군 비율')
plt.ylabel('비율 (%)')

for i, v in enumerate(values):
    plt.text(i, v + 1, f'{v:.1f}%', ha='center')

plt.show()
```

```python
# 파이치트

plt.figure(figsize=(6,6))

plt.pie(
    values,
    labels=labels,
    autopct='%1.1f%%',
    colors=['green','red'],
    startangle=90
)

plt.title('고혈압 위험군 비율')
plt.show()
```

#### 결과를 해석
  - 전체 인구의 절반 이상이 혈압 관리가 필요한 상태
  - 고혈압은 이미 발생한 문제보다, 발생 직전(전단계)이 훨씬 많다 - 잠재적 환자가 많음
  - 혈압 관리를 충분히 해주면 심각한 문제 발생이 일어나지 않음
  - (부모님 건강 관리를 일찍 도와드려야겠다...)

##  [문제 10] 건강 위험 점수 모델링

* 다음 기준으로 건강 위험 점수를 계산하시오.

  * 비만 / 고혈압 / 당뇨 / 흡연 / 음주
* 점수에 따라 그룹을 분류하시오.

  * 건강 / 주의 / 고위험
* 그룹별 인원 비율을 구하고 그래프로 시각화하시오.
* 고위험군 특징을 분석하시오.

#### 비만 / 고혈압 / 당뇨 / 흡연 / 음주 기준 건강 위험 점수 계산

```python
df['건강위험점수'] = 0

# 비만
df.loc[df['BMI'] >= 30, '건강위험점수'] += 1

# 고혈압
df.loc[df['혈압구간'] == '고혈압', '건강위험점수'] += 1

# 당뇨
df.loc[df['혈당구간'] == '당뇨', '건강위험점수'] += 1

# 흡연 (현재흡연 = 3)
df.loc[df['흡연상태'] == 3, '건강위험점수'] += 1

# 음주 (1 = 음주)
df.loc[df['음주여부'] == 1, '건강위험점수'] += 1

df['건강위험점수'].value_counts().sort_index()
```

#### 점수에 따라 그룹 분류

```python
df['위험그룹'] = '고위험'

df.loc[df['건강위험점수'] <= 1, '위험그룹'] = '건강'
df.loc[df['건강위험점수'].between(2, 3), '위험그룹'] = '주의'

df['위험그룹'].value_counts()
```

#### 그래프로 시각화

```python
import matplotlib.pyplot as plt

counts = df['위험그룹'].value_counts()

plt.figure(figsize=(7,5))
plt.bar(counts.index, counts.values, color=['green','orange','red'], edgecolor='black')

plt.title('건강 위험 그룹 분포')
plt.xlabel('위험 그룹')
plt.ylabel('인원 수')

for i, v in enumerate(counts.values):
    plt.text(i, v + 1, f'{v:,}', ha='center')

plt.show()
```

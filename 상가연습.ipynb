```python
# 스트릿 맵

import folium

map_folium = folium.Map(
    location=[37.5706,126.9855], 
    zoom_start=17)
map_folium
```

```python
# 마커와 팝업

# 원래 스트릿 맵
map_folium = folium.Map(location=[37.5706,126.9855], zoom_start=17)

# 팝업
popup = folium.Popup("MBC컴퓨터학원", max_width=200)

# 마커
folium.Marker(location=[37.5706,126.9855], popup=popup).add_to(map_folium)
map_folium
```

```python
# 아이콘 출력
from folium.features import CustomIcon

# 이미지 등록
icon_img = './data/batman.png'
icon = CustomIcon(icon_img, icon_size=(45,45))

seoul_folium = folium.Map(location=[37.56652479270633, 126.9779129116262], zoom_start=14)

popup1 = folium.Popup("서울특별시청", max_width=200)
popup2 = folium.Popup("덕수궁", max_width=200)
popup3 = folium.Popup("서울지방경찰청", max_width=200)
popup4 = folium.Popup("광화문", max_width=200)
popup5 = folium.Popup("영풍문고-종로본점", max_width=200)

# 별 모양 아이콘
folium.Marker(location=[37.56652479270633, 126.9779129116262], popup=popup1, icon=folium.Icon(color="red", icon="star")).add_to(seoul_folium)
# 체크 모양 아이콘
folium.Marker(location=[37.56585320879458, 126.97514572197248], popup=popup2, icon=folium.Icon(color="pink", icon="home")).add_to(seoul_folium)
# 사람 모양 아이콘
folium.Marker(location=[37.57504045678939, 126.97195576938117], popup=popup3, icon=folium.Icon(color="beige", icon="flag'")).add_to(seoul_folium)
# 위에서 만든 배트맨 아이콘
folium.Marker(location=[37.57617987638928, 126.97694467807803], popup=popup4, icon=icon).add_to(seoul_folium)
# 기본 아이콘
folium.Marker(location=[37.56999947440024, 126.98228391170936], popup=popup5).add_to(seoul_folium)

seoul_folium
```

```python
seoul_folium = folium.Map(location=[37.56652479270633, 126.9779129116262], zoom_start=15)

popup1 = folium.Popup("서울특별시청", max_width=200)
# 시청 좌표에 빨간색 원 마커를 표시
folium.CircleMarker(location=[37.56652479270633, 126.9779129116262], popup=popup1, radius=57, color="orange", fill_color="#EC3063").add_to(seoul_folium)
seoul_folium
```

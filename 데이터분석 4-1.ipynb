```python
import seaborn as sns
import pandas as pd

tips = sns.load_dataset('tips')
tips.info()
```

```python
tips.head()
```

```python
import matplotlib.pyplot as plt
#한글폰트설정
plt.rc('font',family='Malgun Gothic')
plt.hist(data=tips, x='total_bill',bins=20)
plt.title("팁")
```

```python
plt.scatter(
    data=tips,
    x='total_bill',
    y='tip'
)
plt.title("팁")
plt.xlabel("요금")
plt.ylabel("팁")
plt.show
```

```python
#종류별로
# 성별
plt.scatter(
    data=tips.query("sex=='Male'"),
    x='total_bill',
    y='tip',
    label="Male"
)
plt.scatter(
    data=tips.query("sex=='Female'"),
    x='total_bill',
    y='tip',
    label="Female"
)
plt.legend() #범례
```

```python
#시간별로
plt.scatter(
    data=tips.query("time=='Dinner'"),
    x='total_bill',
    y='tip',
    label="Dinner"
)
plt.scatter(
    data=tips.query("time=='Lunch'"),
    x='total_bill',
    y='tip',
    label="Lunch"
)
plt.legend()
```

```python
#박스 그래프(분포도 확인 가능)
plt.boxplot(
    [tips.query('sex=="Male"')['tip'],
    tips.query('sex=="Female"')['tip']],
    tick_labels=['Male','Female']
)
plt.show()
```

```python
sns.histplot(data=tips, x='total_bill')
plt.show()
```

```python
sns.kdeplot(data=tips, x='total_bill')
plt.show()
```

```python
sns.kdeplot(data=tips, x='total_bill')
sns.rugplot(data=tips, x='total_bill')
plt.show()
```

```python
sns.countplot(data=tips, x='day')
plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill',y='tip')
plt.show()
```

```python
# 추세선(오차범위)회귀
sns.lmplot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
sns.boxplot(data=tips, x='time', y='total_bill')
plt.show()
```

```python
sns.pairplot(data=tips)
plt.show()
```

```python
tips.query('sex=="Male"')['total_bill'].plot.hist(alpha=.5) #반투명
tips.query('sex=="Female"')['total_bill'].plot.hist(alpha=.5)
```

```python
ax = tips.query('sex=="Male"').plot.scatter(x='total_bill',y='tip')
tips.query('sex=="Female"').plot.scatter(x='total_bill',y='tip', c='pink', ax=ax)
```

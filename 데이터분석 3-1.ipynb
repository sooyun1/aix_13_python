```python
# 연습 문제
import pandas as pd
customers = pd.DataFrame({
    'customers_id': [1,2,3,4],
    'name': ['Alice', 'Bob', 'Charlie', 'David']
})

orders = pd.DataFrame({
    'order_id': [101, 102, 103],
    'customer_id': [1, 2, 5],
    'amount': [25000, 45000, 30000]
})
print(customers)
orders
```

```python
# 기본 레벨
# 답안
# result = pd.merge(customers, orders, on='customer_id', how='inner')

df= customers.merge(
    orders,
    left_on="customers_id",
    right_on="customer_id",
    how="inner").drop(columns=["customer_id"])

print(df)
```

```python
# 문제 2
# 답안
# result = pd.merge(customers, orders, on='customer_id', how='left')

df2 = customers.merge(
    orders,
    left_on="customers_id",
    right_on="customer_id",
    how="left").drop(columns=["customer_id"])

print(df2)
```

```python
# 문제 3
# 답안
# result = pd.merge(customers, orders, on='customer_id', how='right')

df3 = customers.merge(
    orders,
    left_on="customers_id",
    right_on="customer_id",
    how="right").drop(columns=["customer_id"])

print(df3)
```

```python
# 중급 레벨 데이터
jan_fales = pd.DataFrame({
    'product_id': ['P001','P002','P003'],
    'jan_amount': [1000, 2000, 1500]
})

feb_sales = pd.DataFrame({
    'product_id': ['P002','P003','P004'],
    'feb_amount': [2200, 1400, 3000]
})

print(jan_fales)
feb_sales
```

```python
# 중급 레벨 답안
jan_sales = pd.DataFrame({
    'product_id': ['P001', 'P002', 'P003'],
    'jan_amount': [1000, 2000, 1500]
})

feb_sales = pd.DataFrame({
    'product_id': ['P002', 'P003', 'P004'],
    'feb_amount': [2200, 1400, 3000]
})

result = pd.merge(jan_sales, feb_sales, on='product_id', how='outer')
result = result.fillna(0)
result['total'] = result['jan_amount'] + result['feb_amount']
print(result)
```

```python
# 중급 레벨
df = pd.merge(jan_fales, feb_sales, on='product_id', how='outer')

df[['jan_amount', 'feb_amount']] = df[['jan_amount', 'feb_amount']].fillna(0)

# total 컬럼 생성
df['total'] = df['jan_amount'] + df['feb_amount']

print(df)
```

```python
# 고급 레벨 답안
online = pd.DataFrame({
    'customer_id': ['C001', 'C002', 'C003', 'C004'],
    'name': ['홍길동', '김철수', '이영희', '박민수'],
    'email': ['hong@test.com', 'kim@test.com', 'lee@test.com', 'park@test.com'],
    'join_date': ['2023-01-15', '2023-03-20', '2023-06-10', '2023-08-05'],
    'total_purchase': [500000, 1200000, 800000, 0]
})

grades = pd.DataFrame({
    'customer_id': ['C001', 'C002', 'C003', 'C005'],
    'grade': ['Silver', 'Gold', 'Gold', 'Silver'],
    'points': [500, 1500, 1200, 600]
})

result = pd.merge(online, grades, on='customer_id', how='left')
result['grade'] = result['grade'].fillna('Unknown')
print(result)

# 등급별 평균 total_purchase
print(result.groupby('grade')['total_purchase'].mean())
```
